# **Norma wektora, długość wektora w R2**

```python
import numpy as np
np.__version__
```

```python
v1 = np.array([-1, 4])
v1
```

```python
np.linalg.norm(v1) # norma wektora
```

# **Norma euklidesowa w Rn**

```python
v2 =np.array([-10, 10, 13])
np.linalg.norm(v2)
```

# **Odległość dwóch punktów na płaszczyźnie**

```python
p = np.array([3, 0])
q = np.array([0, 4])

np.linalg.norm(p - q)
```

# **Odległość dwóch punktów w przestzreni Rn**

```python
p = np.array([3, 0, 1])
q = np.array([0, 4, 2])

np.linalg.norm(p - q)
```

# **Mnożenie macierzy**

```python
#Przykład_1 liczba kolumny macierzy X(po lewej) jest rózna liczbie wierszy
#macierzy Y(po prawej), mnożenie mozliwe
X = np.array([[5, 3], [3, 9]])
Y = np.array([[1], [-1]])
print(X, '\n')
print(Y)
```

```python
Z = np.dot(X, Y)
Z
```

```python
Z.shape
```

```python
X = np.array([[2, -1, 3], [3, 1, 0]])
Y = np.array([[2, 1, -1], [0, -1, 2], [3, 2, 0]])
print(X, '\n')
print(Y)
```

```python
Z = np.dot(X, Y)
Z
```

```python
Z.shape
```

# **Rozwiązywanie układów równań**

```python
A = np.array([[2, 4], [1, -1]])
print(A)
```

```python
B = np.array([[10], [-1]])
print(B)
```

```python
A_inv = np.linalg.inv(A)
X = np.dot(A_inv, B)
X
```
